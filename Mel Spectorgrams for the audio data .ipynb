```python
import librosa 
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

guitar, sr = librosa.load('Alesis-Fusion-Clean-Guitar-C3.wav')
snare, _ = librosa.load('Alesis-Fusion-Steel-String-Guitar-C4.wav')
```

```python
def to_decibles(signal):
    stft = np.abs(librosa.stft(signal))
    D = librosa.amplitude_to_db(stft, ref= np.max)
    return D

def plot_spec(D,sr,instrument):
    fig,ax = plt.subplots(figsize = (30,10))
    spec = librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='linear', ax=ax)
    ax.set(title = 'Spectrogram of ' + instrument)
    fig.colorbar(spec)
    
```

```python
plot_spec(to_decibles(guitar), sr, 'Guitar')
```

```python
plot_spec(to_decibles(snare),_, 'snare')
```
